# Customer Purchase Data Analysis

## Objective
To clean, analyze, and visualize customer transaction data to extract meaningful business insights.

```python
# Import required libraries for data handling and visualization
import pandas as pd
import matplotlib.pyplot as plt
```

# Upload dataset

```python
# Now load your dataset
df = pd.read_csv("customer_purchase_data (1).csv")

df.head()
```

```python
# Display column names, data types, and non-null values
df.info()

# Display number of rows and columns
df.shape
```

```python
# Check missing values in each column
df.isnull().sum()
```

```python
# Replace missing numerical values with mean of respective column
df = df.fillna(df.mean())
```

```python
# Remove duplicate rows from dataset
df = df.drop_duplicates()

# Reset index after removing duplicates
df = df.reset_index(drop=True)
```

```python
# Generate statistical summary (mean, std, min, max, etc.)
df.describe()
```

```python
# Calculate average values for important columns
print("Average Sales:", df["Sales"].mean())
print("Average Discount:", df["Discount"].mean())
print("Average Transaction Amount:", df["Transaction_Amount"].mean())
```

```python
# Find maximum and minimum values in dataset
print("Max Sales:", df["Sales"].max())
print("Min Sales:", df["Sales"].min())

print("Max Transaction:", df["Transaction_Amount"].max())
print("Min Transaction:", df["Transaction_Amount"].min())
```

```python
# Plot histogram to understand distribution of Sales
plt.hist(df["Sales"], bins=20)

# Add title and labels
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")

# Display graph
plt.show()
```

```python
# Plot scatter graph to analyze relationship between Discount and Sales
plt.scatter(df["Discount"], df["Sales"])

# Add title and labels
plt.title("Discount vs Sales")
plt.xlabel("Discount")
plt.ylabel("Sales")

# Display graph
plt.show()
```

```python
# Plot boxplot to detect outliers in Transaction Amount
plt.boxplot(df["Transaction_Amount"])

# Add title
plt.title("Transaction Amount Spread")

# Display graph
plt.show()
```

```python
# Display correlation matrix between numerical columns
df.corr()
```
